```python
#importar pandas
import pandas as pd                                   # biblioteca para opera√ß√µes com dataframe
!pip
```

```python
# ler arquivo csv com pandas
data = pd.read_csv('iris.csv')
#listar as primeiras observa√ß√µes
data.head()
```

```python
#informa√ß√£o dos dados
data.info()
```

```python
#estat√≠stica geral dos dados
data.describe()
```

```python
#contagem de observa√ß√µes por classe. Verificar o balanceamento de classes para que o algoritmo ajude o nosso problema.
data['species'].value_counts()
```

```python
# Extra√ß√£o dos Atributos / Features. Afinal, eu quero que a m√°quine adivinhe a esp√©cie.
X = data.drop(columns=['species'])
X
```

```python
# Extra√ß√£o do Target / Classe alvo
y = data['species']
y
```

```python
# Cria√ß√£o dos datasets de treino e teste
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state=0)
```

```python
#dimens√µes 
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Cria√ß√£o do modelo √Årvore de decis√£o
from sklearn.tree import DecisionTreeClassifier  #classificador √°rvore de decis√£o
#atribuindo o modelo a clf
clf = DecisionTreeClassifier()
#treinando o modelo com os dados de treino
clf.fit(X_train,y_train)
```

```python
# plotando a √°rvore de decis√£o


import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(20,10))
plot_tree(clf, filled=True, fontsize=12, feature_names=data.columns,
         class_names=data['species'].unique())
plt.show()  #mostrar plot
```

```python
# Acur√°cia da predi√ß√£o
from sklearn.metrics import accuracy_score   #m√©trica acur√°cia
# fazendo a predi√ß√£o do teste
y_pred = clf.predict(X_test)      

# print da acur√°cia sobre a predi√ß√£o dos dados de teste
print("Acur√°cia:", accuracy_score(y_true = y_test, y_pred=y_pred))
```

```python
#imprimir matriz de confus√£o
from sklearn.metrics import confusion_matrix, plot_confusion_matrix  #matrix de confus√£o

#print da matriz de confus√£o
plot_confusion_matrix(clf, X_test, y_test)
```

```python
#testando uma predi√ß√£o
pred = clf.predict ([[5.0, 2.3, 3.3, 1.0]])
print(pred)
```
